```python
 # Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

```python
#Total number of players based on calling the unique on 'SN' using 
#global len to count the number of elements in the returned list and returning as  a dict
player_count ={'Total Player':len(purchase_data['SN'].unique())}

player_count = pd.DataFrame(player_count, index = [0])


player_count
```

```python
#Purchasing Analysis totals

#getting the number of unique items by using the unique func
unique_items = len(purchase_data['Item Name'].unique())

# getting the average purchase price by using the mean func
avg_pp = round(purchase_data['Price'].mean(),2)

# Counting purchases by purchase ID using count func
total_purchases = purchase_data['Purchase ID'].count()

# Getting total revenue by summing price using sum func
total_rev = purchase_data['Price'].sum()

#Creating a dict of variable outputs
purchasing_analysis ={'Number of Unique Items':unique_items, 'Avg Price':avg_pp, 'Number of Purchases':total_purchases, 
                      'Total Rev':total_rev}

#Using dict of purchasing analysis to create a new df
purchasing_analysis = pd.DataFrame(purchasing_analysis, index = [0])



purchasing_analysis

```

```python
gen_dem_df = purchase_data.groupby(['SN','Gender'])['Gender'].size().reset_index(name='Num of Purchases')
gen_dem_df
#Creating a new dataframe based on the user name and gender information to calc gen total and percentage + other calcs
```

```python
# get count of purchase 
gen_dem_cnt_df = purchase_data['Gender'].value_counts().reset_index(name='Num of Purchases')

# Changing the index from gender to number
gen_dem_cnt_df.reset_index(drop=True, inplace=True)

# Renaming Column Name
gen_dem_cnt_df.columns=['Gender','Num of Purchases']

gen_dem_cnt_df
```

```python
# getting breakdown of gender count
gen_count_df = gen_dem_df['Gender'].value_counts().reset_index(name='Gen Cnt')

#Getting total of genders to create percentage and add to df
total_gen = gen_count_df['Gen Cnt'].sum()

#adding percentage to df
gen_count_per = round((gen_count_df['Gen Cnt']/total_gen),3)
gen_count_df['Gen Per'] = gen_count_per

#Renaming columns
gen_count_df.columns=['Gender','Gen Cnt','Gen Per']

gen_count_df.style.format({"Gen Per":"{:.2%}"})



```

```python
# grouping data to get avg price by gender
gen_dem_avg_df = round(purchase_data[['SN','Gender','Price']].groupby('Gender').mean(),2) 

#Changing the index from gender to number
gen_dem_avg_df.reset_index(drop=False, inplace=True)

# Renaming Column Name
gen_dem_avg_df.columns=['Gender','Avg_Price']

gen_dem_avg_df
```

```python
# grouping data to get total purchase value of purchase 
gen_dem_pv_df = purchase_data[['Gender','Price']].groupby('Gender').sum()

gen_dem_pv_df.sort_values('Gender')

#Changing the index from gender to number
gen_dem_pv_df.reset_index(drop=False, inplace=True)

# # Renaming Column Name
gen_dem_pv_df.columns=['Gender','Total Purchase Value']


gen_dem_pv_df
```

```python
# Merging the columns from 3 data frames unsure how to merge in one step so broken out into 2
Gen_PA_df = pd.merge(gen_dem_cnt_df, gen_dem_avg_df, on='Gender',how='left')

#Merging 3rd data frame
Gen_PA_df = pd.merge(Gen_PA_df, gen_dem_pv_df, on='Gender',how='left')


#calculating avg total per person based on TV
Gen_PA_df['Avg Total Per Person'] = round(Gen_PA_df['Total Purchase Value']/gen_count_df['Gen Cnt'],2)

#sorting df
Gen_PA_df.sort_values('Gender', axis=0,ascending=True, inplace=True)

# Styling output
Gen_PA_df.style.format({"Avg_Price":"${:.2f}", "Total Purchase Value":"${:.2f}","Avg Total Per Person":"${:.2f}"})
```

```python
#getting max age for bin
max_age=purchase_data['Age'].max()
print(max_age)

#getting min age for bin
min_age=purchase_data['Age'].min()
print(min_age)
```

```python
# Create the bins for the age groups   
bins = [0, 10, 15, 20, 25, 30, 35, 40, 45]

# Create the names for bins
group_names = ["<10", "11-15", "16-20", "21-25", "26-30","31-35","36-40","40+"]
```

```python
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_names, include_lowest=True)

purchase_data
```

```python
# getting breakdown of gender count step 1 making a smaller df
# I needed to use sn and age group to ensure no one is double counted
age_group_count_df = purchase_data[['Age Group','SN']].value_counts().reset_index(name='Num of Purchases')

# getting breakdown of gender count step 2
age_group_count_df = age_group_count_df['Age Group'].value_counts().reset_index(name='Age Grp Cnt')

# Adding column to calc percetage of diff age groups
age_group_count_df['Percentage of Players'] = round(age_group_count_df['Age Grp Cnt']/576,2)

# #Renaming columns
age_group_count_df.columns=['Age Group','Age Grp Cnt','Percentage of Players']

# Styling output
age_group_count_df.style.format({"Percentage of Players":"{:.2%}"})
```

```python
#Grouping by age group to get purchase count
age_dem_pc_df = purchase_data[['Age Group','Purchase ID']].groupby('Age Group').count()

# Changing the index from gender to number
age_dem_pc_df.reset_index(drop=False, inplace=True)

# Renaming Column Name
age_dem_pc_df.columns=['Age Group','Purchase Count']


age_dem_pc_df
```

```python
#Grouping by age group to get avg purchase price
age_dem_avgpp_df = round(purchase_data[['Age Group','Price']].groupby('Age Group').mean(),2)

# Changing the index from gender to number
age_dem_avgpp_df.reset_index(drop=False, inplace=True)

# Renaming Column Name
age_dem_avgpp_df.columns=['Age Group','Avg Purchase Price']


age_dem_avgpp_df
```

```python
#Grouping by age group to get total purchase value
age_dem_tpv_df = round(purchase_data[['Age Group','Price']].groupby('Age Group').sum(),2)

# Changing the index from gender to number
age_dem_tpv_df.reset_index(drop=False, inplace=True)

# Renaming Column Name
age_dem_tpv_df.columns=['Age Group','Total Purchase Value']


age_dem_tpv_df
```

```python

# Merging the columns from 3 data frames unsure how to merge in one step so broken out into 2
Gen_PA_df = pd.merge(gen_dem_cnt_df, gen_dem_avg_df, on='Gender',how='left')

#Merging 3rd data frame
Gen_PA_df = pd.merge(Gen_PA_df, gen_dem_pv_df, on='Gender',how='left')


#calculating avg total per person based on TV
Gen_PA_df['Avg Total Per Person'] = round(Gen_PA_df['Total Purchase Value']/gen_count_df['Gen Cnt'],2)

#sorting df
Gen_PA_df.sort_values('Gender', axis=0,ascending=True, inplace=True)

# Styling output
Gen_PA_df.style.format({"Avg_Price":"${:.2f}", "Total Purchase Value":"${:.2f}","Avg Total Per Person":"${:.2f}"})


# # grouping data to get avg of purchase by person by age group
# age_dem_APTPP_df = purchase_data[['SN','Age Group','Price']].groupby(['Age Group','SN']).sum()


# # Changing the index from gender to number
# age_dem_APTPP_df.reset_index(drop=False, inplace=True)

# # Renaming Column Name
# age_dem_APTPP_df.columns=['Age Group','SN','Avg Total Purchase']

# age_dem_APTPP_df

# #review NaN
```
